```python
import numpy as np
import matplotlib.pyplot as plt

with open('det_reid_scores.txt', 'r') as f:
    data = f.readlines()
    
def split_record(x):
    a = x.strip('\n').split(' ')
    return [float(a[0]), int(float(a[1])), float(a[2])]
data = np.array(list(map(split_record, data)))
print(data)
```

```python
data_split = []
for i in range(4):
    data_split.append(list(filter(lambda x:x[1] == i and x[2] < 1.0, data)))
# print(data_split)
```

```python
plt.figure()
# plt.scatter(data[:, 0], data[:, 2])
COLOR_COLLECTION = ['r', 'b', 'g', 'y', 'c']
for i in range(4):
    data_i = np.array(data_split[i])
    plt.scatter(data_i[:, 0], data_i[:, 2], label=str(i))
plt.legend(loc='best')
plt.show()
```
